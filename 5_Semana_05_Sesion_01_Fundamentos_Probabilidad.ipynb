## Introducción a la Probabilidad en Ingeniería

En la ingeniería, desde el diseño de estructuras civiles resistentes a sismos hasta la optimización de rendimientos en cultivos agrícolas o el control de calidad en procesos agroindustriales, la incertidumbre es una constante. La probabilidad se convierte en nuestro lenguaje matemático fundamental para cuantificar, analizar y modelar esta incertidumbre, permitiéndonos tomar decisiones informadas y gestionar riesgos.

### Conceptos Fundamentales

Para adentrarnos en el mundo de la probabilidad, es esencial comprender algunos términos clave:

*   **Experimento Aleatorio:** Cualquier proceso cuyo resultado no se puede predecir con certeza antes de su realización, pero del cual conocemos todos sus posibles resultados. Ejemplos incluyen el lanzamiento de un dado, la inspección de la calidad de un material o la medición de la altura de plantas en un campo.
*   **Espacio Muestral ($\Omega$):** Es el conjunto de todos los posibles resultados de un experimento aleatorio. Se denota comúnmente con la letra griega omega mayúscula ($\Omega$). Si inspeccionamos un producto, $\Omega$ podría ser {`Aprobado`, `Defectuoso`}.
*   **Evento ($A$):** Cualquier subconjunto del espacio muestral. Un evento es un resultado o una colección de resultados específicos de interés. Por ejemplo, en el lanzamiento de un dado, el evento "obtener un número par" sería {`2`, `4`, `6`}.

### Ecuaciones Clave de Probabilidad

La probabilidad de un evento se expresa como un número entre 0 y 1 (o 0% y 100%).

#### Probabilidad Teórica (Regla de Laplace)

Cuando todos los resultados de un espacio muestral son igualmente probables, la probabilidad de un evento $A$ se calcula como:

$$P(A) = \frac{\text{Número de casos favorables a A}}{\text{Número total de casos posibles}}
$$

Esta es la definición clásica, útil en situaciones idealizadas o bien definidas, como juegos de azar.

#### Probabilidad Empírica (Frecuencia Relativa)

En la práctica ingenieril, a menudo calculamos probabilidades basándonos en datos observados de experimentos repetidos. La probabilidad empírica de un evento $A$ se estima como:

$$P(A) \approx \frac{\text{Frecuencia con la que ocurre A}}{\text{Número total de observaciones}}
$$

Esta medida se conoce como frecuencia relativa y tiende a la probabilidad teórica a medida que el número de observaciones aumenta.

## **Celda 3: Código Python (Espacio Muestral, Eventos y Probabilidad Empírica)**

````
import pandas as pd
import numpy as np

# Simulamos el registro histórico de inspección de 500 vigas de concreto (Ing. Civil)
# o lotes de producción (Ing. Agroindustrial).
# Resultados posibles (Espacio Muestral): "Aprobado", "Fallo_Menor", "Fallo_Critico"

np.random.seed(42)
espacio_muestral = ['Aprobado', 'Fallo_Menor', 'Fallo_Critico']
# Asignamos una probabilidad teórica subyacente (80% éxito, 15% fallo menor, 5% fallo crítico)
inspecciones = np.random.choice(espacio_muestral, size=500, p=[0.80, 0.15, 0.05])

df_calidad = pd.DataFrame({'Estado_Inspeccion': inspecciones})

# ==========================================
# CÁLCULO DE PROBABILIDAD EMPÍRICA
# ==========================================
print("--- FRECUENCIAS ABSOLUTAS ---")
frecuencias = df_calidad['Estado_Inspeccion'].value_counts()
print(frecuencias)

print("\n--- PROBABILIDAD EMPÍRICA (Frecuencia Relativa) ---")
prob_empirica = df_calidad['Estado_Inspeccion'].value_counts(normalize=True)
print(prob_empirica.round(4))

# Interpretación Ingenieril:
# El "Espacio Muestral" son los 3 estados posibles.
# Un "Evento" puede ser que una viga tenga un 'Fallo_Critico'.
# La Probabilidad Empírica de 'Fallo_Critico' calculada con nuestros datos es cercana a 0.05 (5%),
# calculada simplemente dividiendo los casos favorables entre el total de inspecciones.
````

## Probabilidad Condicional e Independencia en la Ingeniería

Después de calcular las probabilidades empíricas generales, surge una pregunta crucial en ingeniería: ¿la ocurrencia de un evento impacta la probabilidad de otro? Por ejemplo, en el control de calidad, es vital saber si la probabilidad de que una viga de concreto presente un 'Fallo_Critico' depende del 'Proveedor' que la suministró. Esta relación es fundamental para la toma de decisiones y la gestión de riesgos en nuestros proyectos.

### Conceptos Clave

*   **Probabilidad Condicional ($P(A|B)$):** Es la probabilidad de que ocurra el evento $A$ dado que el evento $B$ ya ha ocurrido. Nos permite actualizar nuestra creencia sobre la ocurrencia de $A$ a la luz de nueva información (la ocurrencia de $B$). Es una herramienta esencial para entender cómo los eventos se influyen mutuamente en sistemas complejos, desde la fiabilidad de componentes hasta el éxito de cultivos bajo ciertas condiciones climáticas.

*   **Independencia Estadística:** Dos eventos $A$ y $B$ son estadísticamente independientes si la ocurrencia de uno no afecta la probabilidad de ocurrencia del otro. En el contexto de la ingeniería, esto significa que, por ejemplo, la calidad de un material no depende de su origen si los procesos de ambos proveedores son igualmente robustos. Si hay dependencia, esto indica una relación causal o correlacional que debe ser investigada para optimizar procesos o mitigar riesgos.

### Ecuaciones Clave

#### Probabilidad Condicional
La probabilidad condicional se define formalmente como:

$$P(A|B) = \frac{P(A \cap B)}{P(B)}$$

donde $P(A \cap B)$ es la probabilidad de que ambos eventos $A$ y $B$ ocurran simultáneamente, y $P(B)$ es la probabilidad de que ocurra el evento $B$ (asumiendo $P(B) > 0$).

#### Condición de Independencia
Los eventos $A$ y $B$ son independientes si y solo si se cumple alguna de las siguientes condiciones (son equivalentes):

$$P(A|B) = P(A)$$

O, de manera equivalente (y a menudo más práctica para el cálculo):

$$P(A \cap B) = P(A) \cdot P(B)$$

## **Celda 4: Código Python (Probabilidad Condicional e Independencia)**

````
# Ahora, supongamos que estas inspecciones provienen de dos proveedores distintos
proveedores = np.random.choice(['Proveedor_A', 'Proveedor_B'], size=500, p=[0.6, 0.4])
df_calidad['Proveedor'] = proveedores

# Tabla de contingencia cruzada
tabla_cruzada = pd.crosstab(df_calidad['Proveedor'], df_calidad['Estado_Inspeccion'], margins=True)

print("--- TABLA DE CONTINGENCIA ---")
print(tabla_cruzada)

# Cálculo de Probabilidad Condicional
# P(Fallo_Critico | Proveedor_B) = Casos de Fallo Crítico en Prov_B / Total de Prov_B
total_prov_b = tabla_cruzada.loc['Proveedor_B', 'All']
fallos_prov_b = tabla_cruzada.loc['Proveedor_B', 'Fallo_Critico']

prob_condicional = fallos_prov_b / total_prov_b
print(f"\nProbabilidad de Fallo Crítico DADO que es del Proveedor B: {prob_condicional:.4f}")

# Interpretación: Si la probabilidad condicional es igual a la probabilidad marginal (general),
# decimos que los eventos son INDEPENDIENTES. En la práctica ingenieril, si un proveedor
# tiene una tasa de fallos mucho mayor, la calidad "depende" del proveedor.
````
